```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import RegularGridInterpolator
```

```python
X = np.linspace(-1,1,10)
Y = np.linspace(-1,1,8)

xx,yy = np.meshgrid(X,Y)
```

```python
def f(x,y):
    return x**2 + 3*x*y + 2*y**2

arr = f(xx,yy)
```

```python
interp = RegularGridInterpolator((Y,X),arr)
```

```python
xall = np.linspace(-1,1,101)
yall = np.linspace(-1,1,101)

xxall,yyall = np.meshgrid(xall,yall)

fig,axs = plt.subplots(1,2,sharey=True)

vmin,vmax = -2,4

imf = axs[0].pcolormesh(xxall,yyall,f(xxall,yyall),vmin=vmin,vmax=vmax)
axs[0].set(title='f(x,y)')

imi = axs[1].pcolormesh(xxall,yyall,interp((yyall,xxall)),vmin=vmin,vmax=vmax)
axs[1].set(title='interpolation')

cb = plt.colorbar(imf,ax=axs,location='bottom')
```

```python
fig,ax = plt.subplots()

im = ax.pcolormesh(xxall,yyall,(interp((yyall,xxall))-f(xxall,yyall)),cmap='Greys_r')
ax.plot(xx,yy,ls='',marker='o')
ax.set(title='f - interp')
cb = plt.colorbar(im,ax=ax)
```

```python
fig,ax = plt.subplots()

real = f(xxall,yyall)
guess = interp((yyall,xxall))

im = ax.pcolormesh(xxall,yyall,np.log10(np.abs((guess-real))/real),cmap='Greys')
ax.plot(xx,yy,ls='',marker='o')
ax.set(title='f - interp')
cb = plt.colorbar(im,ax=ax)
```

```python
x = np.linspace(1e-6,1,10)
y = np.linspace(1e-6,1,9)
z = np.linspace(1e-6,1,12)

xx,yy,zz = np.meshgrid(z,y,x,indexing='ij')
def V(x,y,z):
    return x*y*z
data = V(xx,yy,zz)

interp = RegularGridInterpolator((z,y,x),data)

xt = np.pi/10
yt = np.e/10
zt = np.phi/10
print(interp((zt,yt,xt)))
print(V(xt,yt,zt))
```
